## 4. Driving behavior scoring
To score the driving behavior, we use "AvgFuelConsumption" as target of regression model and other driving related features as dependent variables in the regression model.

```python
# Get all driving related features for dependent variable
target = ["AvgFuelConsumption"]
all_col = list(data_driving.columns)    
features = [f for f in all_col if f not in target+["TripID","CustomerID","DriverID",] ]

# Build regression model using Grandient boosting regressor and do the prediction
def reg_pred(data, target, feat=features):
    Y = data.loc[:,target[0]]
    X = data.loc[:,feat]
    GBR = GradientBoostingRegressor(random_state=4, n_estimators=1000, n_iter_no_change=30)
    GBR.fit(X,Y)
    importance = GBR.feature_importances_    
    pred = GBR.predict(X)
    new_data = data.copy()
    new_data["pred"] = pred
    data_score = score_byrank(new_data)
    return importance,data_score

# Rank the trips by predicted fuel conusmption and score the trips according to the ranking
def score_byrank(data,ascending=False):
    temp_data = data.sort_values("pred",ascending=ascending)
    temp_data["score"] = np.arange(len(temp_data))/len(temp_data)
    return temp_data
```

### 4.1 Ranking top 30 important features

```python
# Get feature importances of GBR model and sort the features by importances
def getFeatureImportance(importance,feature_list):
    dict_import=dict()
    for ind,feature in enumerate(feature_list):
        dict_import[feature]=importance[ind]
    dict_import = dict(sorted(dict_import.items(), key=lambda x: np.abs(x[1]),reverse=True)) 
    return dict_import

# PLot the importances of GBR models in each cluster
def plot_importance(importances):
    plt.rcParams["figure.figsize"] = (10,25)
    fig, axs = plt.subplots(len(importances))
    for i in range(len(importances)):
        imp = importances[i]
        axs[i].barh(np.arange(len(list(imp.keys())[0:30])), list(imp.values())[0:30])
        axs[i].set_yticks(np.arange(len(list(imp.keys())[0:30])))# labels=list(importances.keys()))
        axs[i].set_yticklabels(list(imp.keys())[0:30])
        axs[i].invert_yaxis()  # labels read top-to-bottom
        axs[i].set_xlabel('Importance')
        axs[i].set_title("Cluster"+str(i))
    plt.show()
    
plot_importance(importances)
```
